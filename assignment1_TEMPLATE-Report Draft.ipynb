## 1.4 Cleaning Data Error

```python
# Check value of age:
df["age"].value_counts()
```

From value check for age we can see that there are impossible value such as 0, 150 and 160. The expected range of the data is 15 to 22.

Propose solution:
- Modify the 150 to 15
- Modify 160 to 16
- Replace outside range value 0, 10, 11, 12 with 15

```python
# Replace 150 to 15
df["age"].replace(150,15, inplace = True)
# Replace 160 to 16
df["age"].replace(160,16, inplace = True)
# Replace [0, 10, 11, 12] to 15
df["age"].replace([0,10,11,12],15, inplace = True)
# Check the value of age again
df["age"].value_counts()
```

```python
# Check famsize value
df["famsize"].value_counts()
```

```python
# Change "BT","GT" to "GT3" in the data frame
df["famsize"].replace(["BT","GT"],"GT3", inplace = True)
# Display value of famesize
df["famsize"].value_counts()
```
